```python
# QUANTIDADE DE ELEMENTOS UNICOS EM CADA COLUNA:
df.nunique()
```

```python
# VALORES UNICOS POR QUANTIDADE DE COLNAS:
df.nunique(axis=1)
```

```python
# CONTAR VALORES UNICOS - CONCIDERARANDO OS DADOS FALTANTES:
#..Note que na coluna 'Horario' agora temos 60 ao invés de 59 como a cima.
#.. 'DropNa=False' -> Conta até valores nulos ou auxentes.
df.nunique(dropna=False)
```

```python
# DEFINE QUANTIDADE DE VALORES EM ORDEM DECRESCENTE

# Mostra as 5 primeiras e as 5 ultimas de 100
df.nlargest(n = 100, columns = 'Visitante Placar')
```

```python
# MESMO PROCESSO, PORÉM PARA ORDEM CRESCENTE:

df.nsmallest(n = 10, columns = 'Visitante Placar')
```

```python
# PRIMEIRA LINHA DE CADA COLUNA : 

df.loc[df["Visitante Placar"].idxmin()]
```
