```python
# COMO OS ARQUIVOS .CSV ESTÃO (AQUI) NA MESMA PASTA DO .IPYNB
#..POSSO USAR APENAS O PONTO FINAL;
path = r'.'
```

```python
filenames = glob.glob(path + "/*.csv")
```

```python
print(filenames)
```

```python
lista_arquivos = []
```

```python
# Vai percorrer os arquivos .csv na variavel 'filenames' e adicionar 
#..os dois na lista_arquivos

for I in filenames:
    lista_arquivos.append(pd.read_csv(I))
```

```python
# Concatena/Junta os Dois .CSVs  
# "ignore_index=True)"  -> Cria um novo Index para o novo arquivo
#..ignorando os originais


df_full_csv = pd.concat(lista_arquivos, ignore_index=True)
```
