```python
import pandas as pd
```

```python
#  index_col="Visitantes"
# Transformou a coluna Visitante na Coluna Indice
df = pd.read_csv("campeonato-brasileiro-full.csv", index_col="Visitante")
```

```python
df.head()
```

```python
# Transversal dos Dados da Primeira Linha:
df.iloc[0]
```

```python
type(df.iloc[0])
```

```python
# Transversal dos Dados da Ultima Linha:
df.iloc[-1]
```

```python
# Acessando linhas com base no indice:
df.iloc[[1,2,3]]
```

```python
# Slice - Slicing
df.iloc[1:3]         # 1 & 2
```

```python
# All
df.iloc[:]
```

```python
# Slicing
# Acessar Linha X Coluna :

df.iloc[3, 3]
```

```python
df.iloc[0,7]
```

```python
# Primeira Linha do Zero | Coluna 0 até o 2:
# Colunas de ID até DATA:
df.iloc[0,:3]
```

```python
# Linha 2 [Colunas Selecionadas]:
df.iloc[2,[0,2,4,5,6,7]]
```

```python
# Linhas 2 & 3  --- Colunas 4,5,6,7  :
df.iloc[[2,3],[4,5,6,7]]
```

```python
# DIFERENÇAS ENTRE -> LOC & ILOC

# iloc -> Numero das linhas e colnas
# loc  -> Usamos Strings para localizar as informações
```

```python

# TODAS AS LINHAS QUE CONTENHAS A PALAVRA 'Juventude'  (Nome do Time).
#df.loc['Juventude']

# LINHA X COLUNA USANDO 'LOC':

df.loc['Juventude', 'Dia']
```

```python
#  OUTRO EXEMPLO Com Duas Colunas:

df.loc['Juventude', ['Dia', 'Horário']]
```

```python
# CONSULTA MULTIPLA:

df.loc[['Juventude', 'Vasco'], ['Dia', 'Horário']]
```
