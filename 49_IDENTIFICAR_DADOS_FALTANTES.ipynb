```python
import pandas as pd
```

```python
df = pd.read_csv("casas_para_venda_faltantes.csv")
```

### Dados faltantes prejudicam os modelos de Machine Learning 

```python
# Procurar nas colunas (axis=1) e trazer apenas os dados faltantes
df[df.isnull().any(axis=1)]
```

```python
# Substitui os 'Nan' por um valor qualquer:
df_faltante_outro_valor = df.copy()

# Substituiremos os dados faltantes po '-1'  :
df_faltante_outro_valor.fillna(-1, inplace=True)
```

```python
df_faltante_outro_valor.head(25)
```

```python
df.head(25)
```

```python
# Preencher com Média
df_faltantes_media = df.copy()
```

```python
df_faltantes_media.describe()
```

```python
# Preencher valores nulos da coluna 'area' com a média da coluna.
df_faltantes_media["area"].fillna(df_faltantes_media['area'].mean(), inplace=True)
```

```python
# Substitui valores faltantes na coluna 'area' por 116.571429
#.. QUE É O VALOR DA MÉDIA DESTA MESMA COLUNA:
df_faltantes_media.head(26)
```

```python
# COMO FAZER A SUBSTITUIÇÃO EM MAIS DE UMA COLUNA:
df_faltante_varias_colunas = df.copy()
```

```python
# INSERIMOS 'MEDIA' NA COLUNA 'AREA', E MINIMO NA COLUNA 'QUARTOS'
# Dentro do Parenteses FillNa Usamos o esquema Chave / Valor para substituir..
df_faltante_varias_colunas.fillna({ "area":df_faltante_varias_colunas.area.mean(), 
                                  "quartos":df_faltante_varias_colunas.quartos.min() })
```

```python
# PARAMETROS DO MÉTODO:
df_com_fill = df.copy()
```

```python
# PREENCHE OS VALORES FALTANTES COM O VALOR 'VIZINHO' (Anterior)
df_com_fill.fillna(method="ffill", inplace=True)
```

```python
df_com_fill.head()
```

```python
# AGORA VAMOS PREENCHER COM O VALOR POSTERIOR:
df_com_bill = df.copy()
```

```python
df_com_bill.fillna(method="bfill", inplace=True)
```

```python
df_com_bill
```
