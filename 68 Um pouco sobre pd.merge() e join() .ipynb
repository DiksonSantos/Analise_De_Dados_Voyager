# JOIN & MERGE -> Two Tools ONE End

```python
import pandas as pd
```

```python
df_2008 = pd.read_csv('ath_2008.csv')
```

```python
df_2012 = pd.read_csv('ath_2012.csv')
```

```python
# ESTA É UMA FORMA DE JNTAR OS DFs
pd.merge(df_2008, df_2012, how='outer', on='Athlete', suffixes=["_2008", "_2012"], indicator=True))
```

```python
# ESTA É OUTRA FORMA

# O sufixos apontam a origem das colunas

df_2008.join(df_2012, how='outer', lsuffix="_2008", rsuffix='_2012')
```
