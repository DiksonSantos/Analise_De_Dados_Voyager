```python
import warnings
warnings.filterwarnings('ignore')

import pandas as pd
```

```python
df_2008 = pd.read_csv("ath_2008.csv")
```

```python
df_2012 = pd.read_csv("ath_2012.csv")
```

# Vamos usar o Append para juntar os dois .CSVs

```python
# ignore_index -> Isso faz com que ele dê novos index/linhas / continuidade 
#..aos do primeiro DF
# -> Ignore_index Não traz o index original, mas sim dá continuidade ao já existente.
joined = df_2008.append(df_2012, ignore_index=False)


# https://pandas.pydata.org/pandas-docs/stable/reference/api/pandas.DataFrame.append.html
```

```python
display(joined)
```

```python
# Conferindo; Os dois tem as ultimas linhas identicas. Significando que
#..o Df_2012 é agora o final.
joined.tail()
#df_2012.tail()
```

# Agora o metodo de concatenação (Mais Organizado)

```python
# keys=[2008,2012]   -> Deixa bem definido a origem dos dados
# names=['Ano'] -> Nomeia a coluna "Keys" (Que é neste caso o ano)
concatenados = pd.concat([df_2008, df_2012], ignore_index=False, keys=[2008,2012], names=['Ano'])
display(concatenados)


# https://pandas.pydata.org/pandas-docs/stable/reference/api/pandas.concat.html
```

# Agora vamos transformar o Ano numa coluna para referenciar a origem da linha de cada DataFrame

```python
concatenados.reset_index().drop(columns='level_1')
```
