```python
import pandas as pd
```

```python
df_2008 = pd.read_csv('ath_2008.csv')
```

```python
df_2012 = pd.read_csv('ath_2012.csv')
```

```python
# RENOMEANDO UMA DAS COLUNAS
# Athlete Agora será Nomes ....


df_2008.columns = ['Nomes', 'Ouro', 'Prata', 'Bronse']
```

```python
df_2008.columns
```

# Estudo de caso -> O que acontece agora se usarmos o 'on' Sabendo que o nome das colunas dos dois conjuntos não são mais iguais ?

```python
# Vamos ver como fica se Fizermos o Merge dos dois,
#..mesmo que eles tenham nomes diferentes nas colunas

# UM ERRO ACONTECE:

df_full = df_2008.merge(df_2012, how='outer', on='Athlete', suffixes=["_2008", "_2012"], indicator=True)
```

```python
# PRECISAMOS USAR UM PARAMETRO A MAIS PARA QUE O MERGE POSSA ACONTECER
#..MESMO QUE AS COLUNAS TENHAM NOMES DIFERENTES

# Trata-se de "left_on"  && "right_on" 

df_full = df_2008.merge(df_2012, how='outer', left_on='Nomes' ,right_on='Athlete', suffixes=["_2008", "_2012"], indicator=True)
```

```python
# ASSIM ELE SE BASEIA NA POSIÇÃO DAS COLUNAS E NÃO NO NOME PARA FAZER O MERGE
df_full.head(-25)
```

```python
# VAMOS AGORA EXCLUIR DA VISUALIZAÇÃO OS DADOS FALTANTES
#..PARA TERMOS UMA VIZUALIZAÇÃO MAIS ENXUTA:

# -> Preenche os dados faltantes da  coluna 'Nomes' em LEFT com os valores
#.. da coluna 'Athlete' de RIGHT:

# inplace ==True -> Salva as alterações.

df_full.Nomes.fillna(df_full.Athlete, inplace=True)
```

```python
df_full.head(-5)

# -> Nomes Faltantes do lado ESQUERDO foram Preenchidos, do Lado DIREITO
#..Não.
```

```python
# Removendo Colunas:

df_full.drop(["Athlete", "_merge"], axis=1, inplace=True)

df_full


"""COMO NÓS PREENCHEMOS OS NOMES FALTANTES SOMENTE NO LADO/COLUNA 
ESQUERDA. PUDE APAGAR A COLUNA DIREITA (a qual tinha nomes faltantes),
E , A COLUNA _merge QUE AGORA NÃO É MAIS NECESSÁRIA"""
```
