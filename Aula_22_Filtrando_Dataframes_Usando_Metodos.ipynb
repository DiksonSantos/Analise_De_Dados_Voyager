```python
import pandas as pd
```

```python
df = pd.read_csv('campeonato-brasileiro-full.csv')
```

```python
df.head()
```

```python
# VEMOS QUE A COLUNA 'DATA' ESTA COMO 'OBJECT' -> MISTURA LETRAS E NUMEROS:
df.info()
```

```python
# TRANSFORMA A COLUNA EM 'DATETIME':

# Salvamos a Conversão de formato na mesma Coluna:
# Definimos o formato Ano-Mes-Dia assim como mostrado na Tabela
df['Data'] = pd.to_datetime(df['Data'], format='%Y-%m-%d')
```

```python
# AGORA A COLUNA 'DATA' ESTA NO FORMATO DATETIME:
df.info()
```

Com Isso agora podemos acessar a coluna 'Data' de forma diferenciada. Ou seja, o Pandas vai enxergar a informação nesta coluna de forma diferente.

```python
# CRIAMOS UMA NOVA COLUNA -> 'Ano_Jogo' :
# Pegamos apenas o Ano com 'dt.year'
df['Ano_Jogo'] = df['Data'].dt.year
```

```python
# A NOVA COLUNA ESTA LA NO FINAL:
df.head()
```

```python
display(df[['Ano_Jogo']])
```

```python
# PARA CRIAR UM NOVO DATAFRAME COM UMA INFORMAÇÃO ESPECIFICA DO DF ORIGINAL:
df_2002 = df.loc[df.Ano_Jogo == 2002]
df_2002.head()

# pegamos todas as linhas relacionadas com a coluna que contem 2002.
```

```python
# COMO SAR O BETWEEN :
# CRIANDO UM DF COM AS INFORMAÇÕES COM REFERENCIA ENTRE AS DATAS DE 2000 A 2003:

df_bet = df.loc[df.Ano_Jogo.between(2000, 2003, inclusive='both')]

# FONTE:
#  https://pandas.pydata.org/docs/reference/api/pandas.Series.between.html
# Exemplo:
#  df.between(left, right, inclusive='both')
```

```python
# PRIMEIRAS 5 LINHAS DO DF   Ref ao Ano 2000:
df_bet.head()
```

```python
# ULTIMAS 5 LINHAS DO DF   Ref ao ano de 2003:
df_bet.tail()
```

```python
# USANDO O ISIN:

#Lista
intervalo_ano = [2000, 2003]

# Os primeiros 5 Valores estão neste intervalo ?
df.Ano_Jogo.isin(intervalo_ano).head()

# True
```

```python
# Os ultimos 5 ...
df.Ano_Jogo.isin(intervalo_ano).tail()

# False
```

```python
# PEGAR VALORES QUE NÃO ESTÃO NO INTERVALO SUPRA CITADO:

# A diferença esta no simbolo ~ 

df_not_interval = df.loc[~df.Ano_Jogo.isin(intervalo_ano)]
```

```python
# AGORA VEMOS TODOS MENOS 2000 & 2003

df_not_interval.Ano_Jogo.unique()
```
