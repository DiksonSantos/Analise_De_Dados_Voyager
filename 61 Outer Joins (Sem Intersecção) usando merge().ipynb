```python
import pandas as pd
```

```python
df_2008 = pd.read_csv('ath_2008.csv')
```

```python
df_2012 = pd.read_csv('ath_2012.csv')
```

```python
df_2008.head()
```

```python
df_merge_outer = df_2008.merge(df_2012, how='outer', on='Athlete', suffixes=('_2008', '_2012'), indicator=True)

display(df_merge_outer)
```

```python
df_merge_outer._merge.value_counts()
```

```python
# VAMOS ACESSAR A COLUNA '_merge'
# != "both"   -> Trazer todos os valores que nÃ£o sÃ£o comuns.

df_merge_outer_sem_Interc = df_merge_outer.loc[df_merge_outer._merge != "both"]

#display(df_merge_outer_sem_Interc )


# ELIMINAMOS AS 32 INTERCESSÃ•ES
df_merge_outer_sem_Interc._merge.value_counts()
```
