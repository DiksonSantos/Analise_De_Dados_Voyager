```python
import pandas as pd
```

```python
df_TT= pd.read_csv('titanic.csv')
```

```python
df_TT.head(100)
```

```python
# PEGAMOS APENAS DUAS DAS COLUNAS DO DF_TITANIC:
df = df_TT[['Sex', 'Age']]
```

```python
df.head(5)
```

```python
# DESTAS DUAS COLUNAS - 'SEX' & 'AGE' Vamos separar por 'SEX'
#..que vai trazer consigo a 'AGE' associada á informação na linha
df_group = df.groupby('Sex')
type(df_group)
```

```python
# EXIBINDO VERTICALMENTE O SPLIT DAS DUAS COLUNAS 'MALE' & 'FEMALE'
for sex, age in df_group:
    display(sex, age)
```

```python
# Pessoas do sexo masculino e suas idades respectivas
df_group.get_group("male")
```

```python
# Pessoas do sexo feminino + Idades delas:
df_group.get_group("female")
```
