```python
import pandas as pd

import warnings
warnings.filterwarnings('ignore')
```

```python
df_2008 = pd.read_csv("ath_2008_ficticio.csv", index_col="Atleta")
```

```python
df_2012 = pd.read_csv("ath_2012_ficticio.csv", index_col="Atleta")
```

```python
df_2008.head()
#df_2008.tail()
```

```python
df_2012.head()
```

```python
# SOMA O VALOR DAS COLUNAS EQUIVALENTES:

soma = df_2008 + df_2012
display(soma)
```

```python
# OUTRO MEIO DE FAZER A MESMA COISA:

soma_2 = df_2008.add(df_2012)
display(soma_2)
```

```python
# PREENCHER VALORES FALTENTES:

soma_3 = df_2008.add(df_2012, fill_value=0) 
# Larissa agora tem 1 medalha. -> Ele esta somando com a de df_2008
display(soma_3)
```

```python
# SUBTRAÇÃO

df_2012.sub(df_2008)
```
