```python
import pandas as pd
```

```python
df_2008 = pd.read_csv('ath_2008.csv')
```

```python
df_2012 = pd.read_csv('ath_2012.csv')
```

```python
# MERGE DA COLUNA ATLETAS -> VEMOS AS INFORMAÇÕES DELES EM 2008 & 2012

# inner -> Pega apenas a Intercessão

# OU SEJA -> SOMENTE QUANDO UM ATLETA GANHOU MEDALHA(s) NAS DUAS COMPETIÇÕES

df_merge = df_2008.merge(df_2012, how='inner', on='Athlete', suffixes=('_2008', '_2012'), indicator=True)


```
