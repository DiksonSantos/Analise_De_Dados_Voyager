```python
import pandas as pd
```

```python
df_2008 = pd.read_csv('ath_2008.csv')
```

```python
df_2012 = pd.read_csv('ath_2012.csv')
```

```python
df_2008.head()
```

```python
df_2012.head()
```

```python
# INNER -> BOTH -> GANHOU (QQ) MEDALHA NAS DUAS COMPETIÇÕES
# LEFT -> EXCLUI AS VEZES EM QUE O ATLETA GANHOU SOMENTE EM 2012 (RIGHT DF)
#..MOSTRA GANHOS DE 2008 & 2012
# RIGHT -> 2012 & 2008

df_right = df_2008.merge(df_2012, how="right", on='Athlete', suffixes=('_2008','_2012'), indicator=True)
```
