```python
import pandas as pd
```

```python
df = pd.read_csv('campeonato-brasileiro-full.csv')
```

```python
#df.tail(-44)
df.head(60)
```

```python
# Nome de todas as Colunas
df.columns
```

```python
#MOSTRA -> TIPO DE DADO & NUMERO DO INDICE
df.info()
```

```python
# Averiguamos a mesma informação (Dia) escritos de formas distintas:
df['Dia'].unique()
```

```python
# Unificando colunas que já deveriam ser a mesma Ex; Quarta & quarta ...
# Os valores das duas são somados em uma Nova 
df['Dia'].replace(to_replace = ["Quarta-feira", "Terça-feira"], value = ["Quarta-Feira","Terça-Feira"], inplace=True)
```

```python
df['Dia'].unique()
```

```python
# ANteriormente 'Quarta Feira' Eram duas coisas diferentes, Agora elas foram unificadas.
df['Dia'].value_counts()
```

```python
# RENOMEAR COLUNAS:
# Passamos o valor atual e então o Novo Nome/Valor (Mandante_Placar)
df.rename(columns={"Mandante Placar" : "Mandante_Placar"}, inplace=True)
```
