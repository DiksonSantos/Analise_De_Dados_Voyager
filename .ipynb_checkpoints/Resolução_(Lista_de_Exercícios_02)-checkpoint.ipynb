### 1) Calcule para todas as observações a diferença entre as colunas Mandante Placar e Visitante Placar. Faça isso preenchendo as lacunas! 

```python
# 'Pega' Todo o conteudo das colunas 9 & 10:
df.iloc[:, [9,10]]
```

```python
# VALOR MÁXIMO - MIN DA HORIZONTAL (axis = 1) DESAS DUAS COLUNAS 
df.iloc[:, [9,10]].apply(lambda x: x.max() - x.min(), axis = 1)
```

```python
# VALOR MÁXIMO - MIN DA VERTICAL (AXIS =0) DESAS DUAS COLUNAS
df.iloc[:, [9,10]].apply(lambda x: x.max() - x.min(), axis = 0)
```
