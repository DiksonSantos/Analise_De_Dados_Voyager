```python
pd.concat([df_2008, df_2012], ignore_index=False, keys=[2008,2012], names=['Ano'])
```

```python
# FAZENDO O MERGE ENTRE OD DATAFRAMES:
# how=  -> Neste caso é a forma de JOIN que iremos usar
# 
df_2008.merge(df_2012, how='outer', on='Athlete' )
```

```python
# TIPOS DE JOINs:

from IPython.display import Image
Image(filename='Tipos_De_Join.png') 
```

```python
# IDENTIFICANDO A ORIGEM DAS COLUNAS SEGUNDO OS SEUS DFs:

df_2008.merge(df_2012, how='outer', on='Athlete', suffixes=('_2008', '_2012') )
```

```python
# indicator=True  -> MOSTRA SE PERTENSE A PRIMEIRA OU A SEGUNDA TABELA
#.. Left ou Right
OUTER = df_2008.merge(df_2012, how='outer', on='Athlete', suffixes=('_2008', '_2012'), indicator=True)
display(OUTER)
```

```python
# FAZ UM CALCULO COM A QUANTIDADE x ORIGEM:

OUTER._merge.value_counts()
```
