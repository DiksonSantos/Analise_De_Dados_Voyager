```python
import pandas as pd
```

```python
TT = pd.read_csv("titanic.csv")
```

```python
TT.head(10)
```

```python
# VAMOS PEGAR APENAS ALGUNS VALORES DO DF (As primeiras 9 Linhas)
# E, as colunas 4 & 5  -> Sexo & Idade
Slice = TT.iloc[:10, [4,5]]
```

```python
display(Slice)

# Slice agora é um DF com um pedaço do DF Original.
```

```python
# Converter pra uma Lista
Lista_Sex = list(Slice.groupby("Sex"))
display(Lista_Sex)
```

## AGORA SEM A NECESSIDADE DE USAR O SLICE, VAMOS PEGAR ESPECIFICAMENTE AS MESMAS INFORMAÇÕES DO DF ORIGINAL USANDO A LISTA E INDICES DA MESMA:

```python
# NESTE CASO ZERO (Inserido Na primeira posição)
#..ESTA SENDO REPRESENTADO PELO SEXO FEMININO 
# E, UM NA SEGUNDA, MUDA O FORMATO DE EXIBIÇÃO DO DATA FRAME (?).
list(TT.groupby("Sex"))[0][1]
```

```python
# Agora mostra o sexo Masculino, com o mesmo formato de exibição do DF 
#..mais amigavel:
list(TT.groupby("Sex"))[1][1]
```

```python
# MOSTRA A IDADE MÁXIMA E, SE SOBREVIVEU OU NÃO:
TT.groupby("Sex")[["Survived","Age"]].max()
```

```python
# MEDIA DE IDADE:


media_idade = TT.groupby("Sex").mean()
display(media_idade)
```

```python
import matplotlib.pyplot as plt

plt.style.use("seaborn")
```

```python
media_idade.plot(kind="bar", subplots=True, figsize=(7,13), fontsize=13)
plt.show()
```

```python
# Sexo & Idade
media_idadeS = TT.groupby("Sex")["Age"].mean()
media_idadeS.plot(kind="bar", subplots=True, figsize=(4,6), fontsize=13)
plt.show()


# Era entre 27 e 30 anos Fem Mas Respectivamente:
```
