```python
import pandas as pd
```

```python
df = pd.read_csv("campeonato-brasileiro-full.csv")
```

```python
df.head()
```

```python

df["Mandante"].sort_values()
```

```python
# RETORNA DF COMPLETO, PORÉM COM A COLUNA MANDANTE.. EM ORDEM CRESCENTE:
df.sort_values(by='Mandante Placar')
```

```python
# COMEÇANDO DO MAIOR VALOR (CIMA) ATÉ O MENOR (Down):
df.sort_values(by='Mandante Placar', ascending=False)
```

```python
# COM MAIS DE UMA COLUNA:
# Inplace True -> Grava
df.sort_values(by=['Mandante Placar',"Visitante Placar"], ascending=[False, True], inplace=True)
df.head()
```

```python
# VOLTAR DF AO ESTADO ORIGINAL:
df.sort_index(ascending=True, inplace=True)
df.head()
```
