```python
# Mostra todas as ocorrencia "Sabado"  na coluna 'Dia':
sab = df.Dia == 'Sábado'
saturday = sab.count()
print(saturday)
```

```python
# OU
# Mostra False ou True De acordo com cada coluna:
df.Dia == 'Sábado'
```

```python
# Filtro:
quarta = 0
sabado = 0
for I in df.Dia:
    if 'Quarta-feira' in I:
        quarta += 1
    elif I == 'Sábado':
        sabado += 1
        #print(I)

print(f'Quartas Feiras: {quarta}\nSabados: {sabado}')
   
```

```python
# MOSTRA A TABELA INTEIRA, PORÉM NA COLUNA 'Visitante' , Só a ocorrencia
#.. 'Bahia' :

df[df.Visitante == "Bahia"]
```

```python
# iloc  ==  Numeros
# loc   ==  Strings

# TRAZENDO APENAS QUANDO O 'SPORT' JOGOU, E QUAL FOI A ARENA:
# Do lado esquerdo esta o Indice de quando ocorre:

Sport_Arena = df.loc[df.Visitante == 'Sport', 'Arena']
display(Sport_Arena)
```

```python
# MOSTRA A INCIDENCIA DE 'Vasco' na Coluna 'Mandante' , com True ou False:
mandante_vasco = df.Mandante == "Vasco"
mandante_vasco.head()

```

```python
# NO CASO O NOME DA COLUNA CONTÉM ESPAÇO EM BRNACO, ASSIM DEVEMOS NOS
#.. REFERIR A ELA COMO UMA STRING PARA FAZER AS CONSULTAS:


# PEGANDO SOMENTE ONDE O PLACAR FOI MAIOR QUE 2 :
mandante_placar = df["Mandante Placar"] > 2

#mandante_placar.head()

print('Total de Partidas: ',len(mandante_placar))  # 8319

contar = 0
for I in mandante_placar:
    if I == True:
        contar+=1
print('Com mais de 2 gols: ',contar)

# NUM TOTAL DE 8319 PARTIDAS, APENAS 1764 TIVERAM O PLACAR A CIMA DE 2 GOLS.
```

```python
#  &  é o mesmo que 'and'  :
(mandante_vasco & mandante_placar).head()

```

```python
df_vasco_placar = df.loc[mandante_vasco & mandante_placar, ['Data', "Horário","Dia"]]
```
